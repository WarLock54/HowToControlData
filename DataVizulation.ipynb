```python
flight_data = pd.read_csv("/content/flight_delays_test.csv",index_col="Month")
```

```python
flight_data.head(10)
```

```python
plt.figure(figsize=(10,2))
plt.title("Aya göre ")
sns.barplot(x=flight_data.index,y=flight_data["DepTime"])
plt.ylabel("deptime(in minutes)")
```

```python
sns.barplot(x=flight_data.index,y=flight_data["DepTime"])
```

```python
flight_data['DepTime'] = flight_data['DepTime'].astype(str)
```

```python
plt.figure(figsize=(10,3))
plt.title("DepTime Ortalama")
#sns.heatmap(data=flight_data,annot=True)
plt.xlabel("HAVAALANI")
#float degerler heatmap de hata çıkartır.
```

```python
sns.scatterplot(x=flight_data['DepTime'], y=flight_data['Distance'])
```

```python
filtered_data = flight_data[flight_data['Distance'] > 3000]
sns.regplot(x=filtered_data['DepTime'], y=filtered_data['Distance'])
```

```python
sns.scatterplot(x=filtered_data['DepTime'], y=filtered_data['Distance'], hue=filtered_data['DayOfWeek'])
```

```python
scatter_plot = sns.lmplot(x="DepTime", y="Distance", hue="DayOfWeek", data=filtered_data)

# X-ekseni etiketleri için düzenlemeler
plt.xticks(rotation=45, ha='right')

# Mesafe ekleme
plt.subplots_adjust(bottom=0.2)

# Grafik başlığı ekleme
plt.subplots_adjust(top=0.9)
scatter_plot.fig.suptitle('DepTime ve Distance Scatter Plot')

# Grafiği gösterme
plt.show()
```

```python
df=pd.read_csv("catalog.csv")
df
#sns.jointplot(x=filtered_data['DepTime'], y=filtered_data['Distance'], kind="kde")
```

```python
#çekirdek yoğunluğu tahmini (KDE) grafiği. KDE çizimlerine aşina değilseniz
# bunu yumuşatılmış bir histogram olarak düşünebilirsiniz.#
sns.kdeplot(data=df["population"],shade=True)
```

```python
# 2D KDE plot
sns.jointplot(x=df["distance"],y=df["population"],kind="kde")
```

```python
sns.histplot(data=df, x='distance', hue='country_name')
plt.title("Histogram Uzaklık")
```

```python
sns.kdeplot(data=df,x="distance",hue="country_name",shade=True)
plt.title("Dağıtım")
```
